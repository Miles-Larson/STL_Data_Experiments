```python
audio_file = mido.MidiFile() 
audio_track = audio_file.add_track('Amazing')

def new_selections():
    global choose_scale, note_options
    choose_scale = random.choice(list(scales.keys()))
    note_options = []
    for i in scales[choose_scale]:
        for x in notes[i]:
            note_options.append(x)
    window_floor = random.randrange(start=1,stop=128)
    window_ceiling = random.randrange(start=window_floor,stop=128)
    note_options = [int(y) for y in note_options  if int(y) > window_floor and int(y) < window_ceiling ]
    if len(note_options) == 0:
        new_selections()
    else:
        print(note_options)
    return note_options

def select_notes_to_play(allowable_notes, phrase_length):
    for i in range(phrase_length):
        random.choice(note_options)
    return 1



def build_midi_section(phrase:object):
    for i in range(phrase.phrase_length):
        midi_message = mido.Message('note_on')
        #midi_message = midi_message.copy(note=(random.randrange(start=0,stop=128)))
        midi_message = midi_message.copy(note=(phrase.play_notes[i]))
        midi_message = midi_message.copy(channel=(phrase.channel_l[i]))
        midi_message = midi_message.copy(velocity=(phrase.velocity_l[i]))
        midi_message = midi_message.copy(time=(phrase.time_l[i]))
        audio_track.append(midi_message)
    return audio_track

class phrase:
    def __init__(self, phrase_length:int, note_selection:list ):
        self.phrase_length:int = phrase_length
        self.note_selection = note_selection
        self.play_notes = []
        self.velocity_l = []
        self.time_l = []
        self.channel_l = []

        for i in range(self.phrase_length):
            self.play_notes.append(random.choice(self.note_selection))
            self.velocity_l.append(random.randrange(start=0,stop=100))
            self.time_l.append(random.randrange(start=0,stop=140))
            self.channel_l.append(random.randrange(start=0,stop=15))
        
    def new_volicity(self):
        self.velocity_l = []
        for i in range(self.phrase_length):
            self.velocity_l.append(random.randrange(start=0,stop=100))
        return self

    def new_time(self):
        self.time_l = []
        for i in range(self.phrase_length):
            self.time_l.append(random.randrange(start=0,stop=140))
        return self


        
    def new_channel(self):
        self.channel_l = []
        for i in range(self.phrase_length):
            self.channel_l.append(random.randrange(start=0,stop=15))
        return self

```
